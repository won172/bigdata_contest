```python
## 음식점 업종 매핑

mapping = {
    '한식': ['한식', '백반/가정식', '분식', '기사식당', '한정식', '떡/한과'],
    '일식': ['일식', '이자카야', '초밥', '덮밥', '우동'],
    '중식': ['중식'],
    '서양식': ['양식', '스테이크', '피자', '햄버거'],
    '기타 아시아/세계음식': ['동남아/인도음식', '기타세계요리'],
    '음료/디저트/베이커리': ['커피', '카페', '아이스크림/빙수', '주스', '차', '베이커리', '마카롱', '도너츠', '탕후루', '샌드위치', '와플'],
    '주점/유흥': ['요리주점', '호프/맥주', '포장마차', '민속주점', '룸살롱/단란주점', '유흥주점', '와인바'],
    '식료품/소매': ['식료품', '축산물', '수산물', '농산물', '청과물', '미곡상', '건어물', '담배', '주류', '와인샵', '반찬', '유제품'],
    '간편식/가공식품': ['도시락', '식품 제조', '건강식품', '인삼제품', '건강원', '떡/한과 제조'],
    '기타': ['꼬치구이', '구내식당/푸드코트']
}
```

```python
# 매핑 함수 작성
def map_category(category):
    for group, items in mapping.items():
        for item in items:
            if item in category:
                return group
    return '기타' 

df_m
df_m['업종_대분류'] = df_m['HPSN_MCT_ZCD_NM'].apply(map_category)

print('\n전체 대분류별 개수:\n', df_m['업종_대분류'].value_counts())
```

```python
df_grouped = df_m.groupby(['TA_YM', '업종_대분류'])[['M12_MAL_1020_RAT','M12_MAL_30_RAT', 'M12_MAL_40_RAT', 'M12_MAL_50_RAT', 'M12_MAL_60_RAT',
                                                 'M12_FME_1020_RAT','M12_FME_30_RAT', 'M12_FME_40_RAT', 'M12_FME_50_RAT', 'M12_FME_60_RAT' ]].mean().reset_index()
```

```python
df_grouped
```

```python
# 연령대 칼럼
age_cols = ['M12_MAL_1020_RAT','M12_MAL_30_RAT', 'M12_MAL_40_RAT', 'M12_MAL_50_RAT', 'M12_MAL_60_RAT',
            'M12_FME_1020_RAT','M12_FME_30_RAT', 'M12_FME_40_RAT', 'M12_FME_50_RAT', 'M12_FME_60_RAT' ]


# 매핑
gender_map = {'MAL': '남성', 'FME': '여성'}
age_map = {'1020': '10-20대', '30': '30대', '40': '40대', '50': '50대', '60': '60대 이상'}

# 업종별로 시계열 플롯
for age_col in age_cols:
    parts = age_col.split('_')
    gender_abbr = parts[1]
    age_str = parts[2]
    
    gender = gender_map.get(gender_abbr, '알 수 없음')
    age_group = age_map.get(age_str, '알 수 없음')
    
    title_text = f"업종별 {gender} {age_group} 비율 변화 추이"
    plt.figure(figsize=(12, 6))
    for industry, group in df_grouped.groupby("업종_대분류"):
        plt.plot(group["TA_YM"], group[age_col], label=industry)

    plt.title(title_text)
    plt.xlabel("시점")
    plt.ylabel("비율 (%)")
    plt.legend()
    plt.show()
```
